```python
def remove_punctuation_and_digits(input_string):

    lowercase_string = input_string.lower()


    for punctuation in string.punctuation:
        lowercase_string = lowercase_string.replace(punctuation, '')


    for digit in string.digits:
        lowercase_string = lowercase_string.replace(digit, '')

    return lowercase_string
```

```python
df['clean'] = df['Text'].apply(remove_punctuation_and_digits)
```
