```python
import pandas as pd
import numpy as np
import emoji
import re
```

```python
df = pd.read_csv('/content/drive/MyDrive/lab programs/smiling_face_with_heart-eyes.csv')
```

```python
# function to replace the emoji with the exact meaning
def replace_emojis(text):
    textwithmeanings = emoji.demojize(text)  # Replace emojis with their meanings
    return textwithmeanings
```

```python
# display the results of the above code
df['without_emoji']=df["Text"].apply(replace_emojis)
```

```python
df
```

```python
def remove_urls(text):
    # Remove URLs using regular expression
    pattern = re.compile(r'https?://\S+|www\.\S+')
    text_without_urls = pattern.sub('',text) #replaces with an empty string
    return text_without_urls
```

```python
# display the results of the above code
df['without_url']=df["without_emoji"].apply(remove_urls)
```
